```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
x = np.arange(-8,8.1,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,'g--',label='sin(x)')
plt.plot(x,y2,'o',color='r',label='cos(x)')
plt.legend(loc='best')
plt.xlim([-10,10])
plt.ylim([-2,2])
```
