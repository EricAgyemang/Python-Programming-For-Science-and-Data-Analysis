```python
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

class ISUFall2018Calendar(AbstractHolidayCalendar):
    rules = [
        Holiday('Thanksgiving', month=11, day=19),
        Holiday("Thanksgiving", month=11, day=20),
        Holiday('Thanksgiving', month=11, day=21),
        Holiday('Thanksgiving', month=11, day=22),
        Holiday('Thanksgiving', month=11, day=23),
    ]

isu = CustomBusinessDay(calendar=ISUFall2018Calendar())
calendar = pd.date_range(start='8/20/2018',end='12/14/2018',freq=isu)
calendar
```

```python
print("There are %d weekdays in Fall 2018."%(calendar.size))
```

```python
it166 = CustomBusinessDay(weekmask = 'Mon Wed', calendar=ISUFall2018Calendar())
IT166 = pd.date_range(start='8/20/2018',end='12/14/2018',freq=it166)
IT166
```

```python
removes = []
for each in IT166:
    if each.month == 12 and each.day in range(10,15):
        removes.append(each)
IT166 = IT166.drop(removes)
print("There are %d lectures of IT166 in Fall 2018."%(IT166.size))
```
